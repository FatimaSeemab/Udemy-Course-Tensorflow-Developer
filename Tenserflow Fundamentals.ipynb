## Manipulating tensors (tensor operations)
**Basic Operations**
`+`,`-`,`*`,`/`

```python
tensor_op = tf.constant([[10,7],[2,5]])
tensor_op + 10
```

```python
# Original tensor is unchanged
tensor_op
```

```python
tensor_op - 10
```

```python
tensor_op * 10
```

```python
## We can use tensorflow built in function
tf.multiply(tensor_op,10)
```

```python
tensor_op
```

**Matrix Multiplication**

```python
tf.matmul(tensor_op,tensor_op)
```

```python
tensor_op*tensor_op
```

```python
# Matrix multiplication with python operator @
tensor_op@tensor_op
```

```python
# CReate a tensor of 3,2
X=tf.constant([[1,2],[3,4],[5,6]])
# Create a tensor of 3,2
Y=tf.constant([[5,6],[7,8],[9,10]])
X@tf.transpose(Y)
```

```python
X@tf.reshape(Y, shape=(2,3))
```

**The dot product**
Dot product can be done by using two operations

1.   Matmul
2.   tensordot


```python
X,Y
```

```python
tf.tensordot(X,tf.transpose(Y),axes=1)
```
